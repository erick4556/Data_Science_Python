```python
#Matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(2,9,10) #Arreglo de 10 elementos del 2 al 9
x
```

```python
y = x**2 #x elevada a la 2
```

```python
y
```

```python
plt.plot(x,y,'red') #Generar el gráfico
plt.title("Gráfico")
plt.xlabel("Eje x") #Ajustar el texto para el eje x
plt.ylabel("Eje y") #Ajustar el texto para el eje x
```

Generar gráficos en el mismo espacio

```python
plt.subplot(1,2,1) #1 fila, 2 columna y la posición1
plt.plot(x,y,'g')
plt.subplot(1,2,2)
plt.plot(y,x,'blue')
```

Gráficos con programación orientada a objetos

```python
figure = plt.figure()
graphic = figure.add_axes([0.1,0.1,0.8,0.8]) #Rellena con un 80% (0.8) el ancho al igual que el alto
graphic.plot(x,y,'g')
```

Generar 2 gráficos dentro de una figura

```python
figure2 = plt.figure()
graphic1 = figure2.add_axes([0.1,0.1, 0.9, 0.9]) 
graphic1.plot(x,y,'r')

graphic2 = figure2.add_axes([0.2, 0.5, 0.3, 0.4]) #Esta separado al 20% (0.2) de la izquierda 50% (0.5) de abajo y tiene un 30% (0.3) de ancho y 40% (0.4) de altura
graphic2.plot(x,y,'b')
```

Multigráficos

```python
#Gráfico de 4 subgráficos
figure, graphics = plt.subplots(nrows=2, ncols=2) #Se ordena como los arrays empezando desde 0
graphics[0][0].plot(x,y,'blue') #Rellena la fila 0 y columna 0
graphics[0][1].plot(x,y,'red')
graphics[1][0].plot(x,y,'y')
graphics[1][1].plot(x,y,'g')

graphics[0][0].set_title("Blue")
graphics[0][1].set_title("Red")
graphics[1][0].set_title("Yellow")
graphics[1][1].set_title("Green")
```

Tamaño de una figura y un gráfico dentro de una figura

```python
figure = plt.figure(figsize=(8,4)) #figsize es para ajustar el ancho y alto en pulgada de la figura
graphic = figure.add_axes([0.1, 0.1, 0.9, 0.9])
graphic.plot(x,y,'r')
```

Dos gráficos en la misma figura

```python
figure = plt.figure()
graphics = figure.add_axes([0.1,0.1,0.9,0.9])
graphics.plot(x, x**2, label="Eje y es x al cuadrado")
graphics.plot(x, x**3, label="Eje y es x al cubo")
graphics.legend(loc=0) #loc=0 es para que busque la mejor posición para poner la leyenda
```
