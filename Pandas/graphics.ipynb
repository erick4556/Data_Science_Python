```python
#Gráficos
import numpy as np
import pandas as pd
```

```python
dataframe = pd.DataFrame(np.random.randint(200, size =(50,4)), columns=['a','b', 'c', 'd']) #Data frame de 200 números randoms de 50 por 4
```

```python
dataframe
```

```python
dataframe['a'].hist() #Crear un histograma de la columna a
```

```python
dataframe['a'].hist(bins=30) #Crear un histograma de la columna b pero con 30 columnas
```

```python
dataframe['a'].plot.hist()
```

```python
dataframe.plot.area(alpha=0.3) #Gráfico de tipo área, parámetro alpha para ajustar el color
```

```python
dataframe.plot.bar() #Gráfico de barra
```

```python
dataframe.plot.bar(stacked=True) #Gráfico de barra con acumulación de valores
```

```python
dataframe.plot.scatter(x='a', y='b') #Gráfico de tipo scatter, donde se le paso los ejes con las columnas
```

```python
dataframe.plot.scatter(x='a', y='b', c='c', cmap='coolwarm') #Gráfico de tipo scatter, con 3 columnas, cmpa para añadir una paleta de colores
```

```python
dataframe.plot.box() #Gráfico de caja
```

```python
dataframe.plot.hexbin(x='a', y='b', gridsize=12) #Gráfico tipo hexadecimal. Puntos más oscuros son de mayor número de elementos
```

```python
dataframe.plot.kde() #Gráfico de tipo kde
```

```python
dataframe.plot.density() #Gráfico de tipo kde de otra forma
```
