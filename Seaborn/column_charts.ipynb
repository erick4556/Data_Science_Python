Gráficos para categorías

```python
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head(10)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips) #Gráfico en base al sexo
```

```python
sns.countplot(x='sex', data=tips) #Gráfico para contar el número de personas por sexo
```

Diagrama de caja

```python
sns.boxplot(x='day', y='total_bill', data=tips) #Gráfico para comparar un gráfico de columna categoría por una no numerica
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker') #hue para filtrar por algun campo
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', split=True) #split para juntar las diferencias entre smoker
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex') #Gráfico para comparar los días con los de la factura
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex') #Gráfico para comparar los días con los de la factura y evitar que se solapen
```

```python
#Juntar varios gráficos
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips)
```
