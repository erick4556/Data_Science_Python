Gráficos de cuadrícula {grids}: PairGrid y FacetGrid

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flowers = sns.load_dataset("iris")
flowers.head()
```

```python
sns.pairplot(flowers) #Construir gráficos de todas las columnas del dataset, hace gráficos 2 a 2
```

```python
graphics = sns.PairGrid(flowers)
graphics.map(plt.scatter) #Gráficos de tipo scatter
```

PairGrid

```python
graphics = sns.PairGrid(flowers)
graphics.map_diag(sns.distplot)
graphics.map_upper(plt.scatter)
graphics.map_lower(sns.kdeplot)
```

FacetGrid

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['sex'].unique() #Cuantos elementos diferentes en el dataset
```

```python
graphics2 = sns.FacetGrid(data = tips, col='time', row='smoker')
graphics2.map(sns.distplot, 'total_bill')
```

```python
graphics2 = sns.FacetGrid(data = tips, col='time', row='smoker')
graphics2.map(plt.scatter, 'total_bill', 'tip') #Como el scatter compara 2 variables de tipo numérico, tengo que pasarle otra variable
```
