```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips') #Cargar el dataset por defecto "tips" para hacer pruebas
```

```python
tips.head(10) #head se le pasa el n√∫mero de registros mostrar
```

```python
sns.distplot(tips['total_bill']) #Crear un gr√°fico de tipo histograma de la columna total_bill
```

```python
sns.distplot(tips['total_bill'], bins = 20) #bins es el n√∫mero de barra del histograma
```

```python
sns.distplot(tips['tip'], bins = 20)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips) #Comparar variables por medio de gr√°fico histogramas
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex') #Comparar variables por medio de gr√°fico hexadecimal (kind=hex)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg') #Comparar variables por medio de gr√°fico regresi√≥n lineal (kind=reg)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde') #Comparar variables por medio de gr√°fico de densidad (kind=kde)
```

```python
sns.pairplot(tips) #Comparar todas las columnas
```

```python
sns.pairplot(tips, hue='sex') #Comparar con las columnas, hue se le pasa la columna con texto a comparar
```

```python
sns.pairplot(tips, hue='size')
```

```python
sns.rugplot(tips['total_bill']) #Mostrar por medio de un gr√°fico, medinate una raya cada uno de los datos que hay en el dataset
```

```python
sns.rugplot(tips['tip'])
```
