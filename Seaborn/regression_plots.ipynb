Gráficos de regresión (Regressión Plots)

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex') #hue=sex para diferenciar entre hombre y mujeres
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','x']) #markers para cambiar los marcadores
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','x'], scatter_kws = {'s': 80}) #scatter_kws para modificar el tamaño
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day') #Separa el gráfico con el parametro col y se pasa la columna
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', row='sex') #Con row se puede añadir las filas
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex') #Utilizar hue para hacer las diferencia entre el sexo
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, aspect=0.8, size=10) #Amplear el gráfico, aspect para el estrecho y size para el tamaño general
```
